**ALGORITMA PEMROGRAMAN MEMAHAMI DASAR-DASAR DAN IMPLEMENTASINYA**

Algoritma adalah urutan langkah-langkah logis yang digunakan untuk menyelesaikan masalah. Setiap langkah harus jelas, dapat dieksekusi, dan merupakan dasar dari pemrograman komputer.

Definisi Program/Pemrograman
Pemrograman adalah kumpulan instruksi-instruksi yang biasanya disebut source code yang dibuat oleh programmer (pembuat program).

**Paradigma Pemrograman**

1. Pemrograman Prosedural

Berdasarkan urutan-urutan, sekuensial.
Program adalah suatu rangkaian prosedur untuk memanipulasi data.
Prosedur merupakan kumpulan instruksi yang dikerjakan secara berurutan.
Harus mengingat prosedur mana yang sudah dipanggil dan apa yang sudah diubah.

2. Pemrograman Fungsional

Berdasarkan teori fungsi matematika.
Fungsi merupakan dasar utama program.

3. Pemrograman Terstruktur

Secara berurutan dan terstruktur.
Program dapat dibagi-bagi menjadi prosedur dan fungsi.

4. Pemrograman Modular

Membentuk banyak modul.
Modul merupakan kumpulan dari prosedur dan fungsi yang berdiri sendiri.
Sebuah program dapat merupakan kumpulan modul-modul.

5. Pemrograman Berorientasi Obyek

Berdasarkan prinsip objek, di mana objek memiliki data/variabel/properti dan metode/prosedur yang dapat dimanipulasi.

6. Pemrograman Berorientasi Fungsi

Berfokus pada fungsi tertentu saja.
Sangat tergantung pada tujuan pembuatan bahasa pemrograman ini.

7. Pemrograman Deklaratif

Mendeskripsikan suatu masalah dengan pernyataan daripada memecahkan masalah dengan implementasi algoritma.

**Sejarah Algoritma**

Istilah "algoritma" berasal dari nama Abu Jaâ€™far Mohammed Ibn Musa alKhwarizmi, seorang ilmuwan Persia. Konsep algoritma sudah dikenal sejak zaman kuno, seperti Algoritma Euclid untuk menemukan pembagi terbesar dua bilangan.

Perkembangan algoritma berlanjut dengan mesin komputasi oleh Charles Babbage dan kolaborasi Ada Lovelace, hingga era modern dengan kontribusi dari Alan Turing dan John von Neumann.

1. **Algoritma di Dunia Kuno**:

- Meskipun istilah "algoritma" baru muncul pada abad ke-9, konsep pemecahan masalah melalui langkah-langkah sistematis sudah ada sejak zaman kuno.

- Tabel perhitungan Babilonia (sekitar 1600 SM) menunjukkan teknik-teknik komputasi dasar.

- Euclid (abad ke-3 SM) menyusun algoritma yang terkenal, Algoritma Euclid, untuk menemukan pembagi terbesar dua bilangan.

- Archimedes juga menggunakan algoritma dalam karyanya tentang perhitungan volume dan luas.


2. **Revolusi Algoritma dan Komputer**:

- Revolusi algoritma dimulai dengan penemuan mesin komputasi modern. Charles Babbage, matematikawan Inggris pada abad ke-19, merancang Mesin Analitik, yang dianggap sebagai komputer mekanik pertama. Mesin ini dimaksudkan untuk menjalankan algoritma secara otomatis, meskipun tidak pernah selesai dibuat.

- Ada Lovelace, kolaborator Babbage, dianggap sebagai programmer pertama di dunia. Ia menulis deskripsi rinci tentang bagaimana mesin tersebut dapat menjalankan serangkaian instruksi untuk menyelesaikan masalah, yang kini dikenal sebagai "algoritma pemrograman."


3. **Algoritma di Era Komputasi Modern**:

- Pada abad ke-20, perkembangan ilmu komputer menciptakan kebutuhan besar akan algoritma yang lebih kompleks.

- Alan Turing, yang dikenal sebagai bapak ilmu komputer modern, mengembangkan model komputasi teoretis yang disebut Mesin Turing pada tahun 1936. Turing mengkonseptualisasikan ide bahwa setiap proses komputasi dapat dipecah menjadi serangkaian langkah algoritmis.

- John von Neumann juga memberikan kontribusi besar dalam pengembangan komputer modern dengan merancang arsitektur komputasi yang mendukung eksekusi algoritma secara efisien.


4. **Perkembangan Algoritma di Dunia Digital**:

- Seiring berkembangnya teknologi komputer, algoritma menjadi pusat dari berbagai aplikasi komputasi modern.

- Algoritma Sorting seperti Bubble Sort, Quick Sort, dan Merge Sort mulai dikembangkan untuk mengatur data secara efisien.

- Pada 1950-an dan 1960-an, Donald Knuth menerbitkan buku "The Art of Computer Programming", yang menjadi referensi penting dalam dunia algoritma. Dia memperkenalkan banyak metode dan strategi optimasi algoritma.

- Algoritma Kriptografi, seperti RSA (Rivest-Shamir-Adleman), dikembangkan pada 1970-an untuk mengamankan data di dunia digital, yang menjadi dasar dari komunikasi yang aman saat ini.


5. **Algoritma di Era Kecerdasan Buatan(AI)**:

- Pada abad ke-21, algoritma memegang peranan penting dalam perkembangan kecerdasan buatan (AI), machine learning, dan big data.

- Algoritma Neural Networks dan Deep Learning memungkinkan komputer untuk "belajar" dari data dan membuat keputusan yang kompleks.

- Penerapan algoritma pada search engines (Google), media sosial, hingga otomasi di berbagai industri membuat algoritma menjadi bagian tak terpisahkan dari kehidupan modern.



6. **Masa Depan Algoritma**:

- Di masa depan, algoritma terus berkembang dengan peningkatan dalam komputasi kuantum, pengenalan pola, dan optimasi sistem kompleks.

- Algoritma tidak hanya berperan dalam komputerisasi tetapi juga dalam pemecahan masalah di bidang lain seperti medis, keuangan, dan robotik.

**Jenis Proses Algoritma**

1.Sequence Process: Instruksi dilakukan secara berurutan.

2.Selection Process: Instruksi dilakukan jika memenuhi kondisi tertentu.

3.Iteration Process: Instruksi dilakukan berulang kali selama kondisi terpenuhi.

4.Concurrent Process: Beberapa instruksi dijalankan secara bersamaan.

**Langkah Penyusunan Algoritma**:

1. Identifikasi masalah.


2. Tentukan input dan output yang diinginkan.


3. Definisikan langkah-langkah secara logis dan terurut.


4. Periksa solusi untuk memastikan hasil yang sesuai dengan tujuan.
